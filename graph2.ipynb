```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./datos2.csv")
df.tail()
```

```python
plt.plot(df['inversion'], label='Inversion')
plt.plot(df['displacement'], label='Displacement')
plt.plot(df['exchange'], label='Exchange')
plt.plot(df['insertion'], label='Insertion')
plt.plot(df['inverted_exchange'], label='Inverted Exchange')
plt.plot(df['inverted_displacement'], label='Inverted Displacement')
plt.ylabel('Minimum distance in km')
plt.xlabel('Generations')
plt.title('Performance of genetic algortithms')
plt.legend()
```
